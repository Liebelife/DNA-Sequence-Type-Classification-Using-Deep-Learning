## Functions

```python
from tokenizers.pre_tokenizers import Whitespace

def get_tokenizer(FilePath, SaveTo, TokenizerType, VocabSize):

    path = os.path.abspath(os.path.join(SaveTo, f'{VocabSize}'))
    
    # create a folder if it doesn't already exist
    !mkdir -p {path}
    
    tokenizer = TokenizerType
    tokenizer.pre_tokenizer = Whitespace()
    trainer = UnigramTrainer(vocab_size=VocabSize, min_frequency=2,\
                        special_tokens=[
                                        "<s>",
                                        "<pad>",
                                        "</s>",
                                        "<unk>",
                                        "<mask>",
                                    ])
    tokenizer.train([FilePath], trainer) # training the tokenzier

    tokenizer.save_model(path)
    print(tokenizer)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import Unigram
 
tokenizer = Tokenizer(Unigram())

from tokenizers.trainers import UnigramTrainer
trainer = UnigramTrainer(vocab_size=15000, min_frequency=2,\
                         unk_token="<unk>",\
                            special_tokens=[
                                            "<s>",
                                            "<pad>",
                                            "</s>",
                                            "<unk>",
                                            "<mask>",
                                        ])
tokenizer.train(["test.raw"], trainer)
tokenizer.save("./tokenizer-trained.json")
```

```python
tokenizer = Tokenizer.from_file("./tokenizer-trained.json")
```
